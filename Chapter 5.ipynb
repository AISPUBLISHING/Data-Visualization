# Chapter 5: Advanced Plotting with Seaborn Library

### 5.1. Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [10,8]

tips_data = sns.load_dataset('tips')

tips_data.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips_data)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips_data, color = 'r')
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips_data, color = 'b', marker = 'x')
```

### 5.2. Styling Seaborn Plots

```python
sns.set_style('darkgrid')
sns.scatterplot(x="total_bill", y="tip", data=tips_data, color = 'b', marker = 'x')
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x="total_bill", y="tip", data=tips_data, color = 'b', marker = 'x')
```

```python
sns.set_context('poster')
sns.scatterplot(x="total_bill", y="tip", data=tips_data, color = 'b', marker = 'x')
```

### 5.3. Heatmaps

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [8,6]
sns.set_style("darkgrid")

titanic_data = sns.load_dataset('titanic')

titanic_data.head()
```

```python
titanic_data.corr()
```

```python
plt.rcParams["figure.figsize"] = [10,8]
corr_values = titanic_data.corr()
sns.heatmap(corr_values, annot= True)
```

```python
plt.rcParams["figure.figsize"] = [10,8]
corr_values = titanic_data.corr()
ax = sns.heatmap(corr_values, annot= True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
plt.rcParams["figure.figsize"] = [10,8]
corr_values = titanic_data.corr()
ax = sns.heatmap(corr_values, annot= True, cmap = 'coolwarm')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [10,8]

flights_data = sns.load_dataset('flights')

flights_data.head()
```

```python

flights_data.tail()
```

```python
flights_data_pivot =flights_data.pivot_table(index='month', columns='year', values='passengers')
ax = sns.heatmap(flights_data_pivot,  cmap = 'coolwarm', linecolor='black', linewidth=1)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

### 5.5. Pair Grids

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [10,8]

tips_data = sns.load_dataset('tips')

```

```python
sns.pairplot(tips_data)
```

```python
sns.PairGrid(tips_data)
```

```python
pgrids = sns.PairGrid(tips_data)
pgrids.map(plt.scatter)
```

```python
pgrids = sns.PairGrid(tips_data)

pgrids.map_diag(sns.distplot)
pgrids.map_upper(sns.kdeplot)
pgrids.map_lower(plt.scatter)
```

### 5.6. Facet Grids

```python
fgrid = sns.FacetGrid(data=tips_data, col='sex', row='time')
fgrid.map(sns.distplot, 'total_bill')
```

```python
fgrid = sns.FacetGrid(data=tips_data, col='sex', row='time')
fgrid.map(plt.scatter, 'total_bill', 'size')
```

### 5.7. Regression Plots

```python
sns.lmplot(x='total_bill', y='tip', data =tips_data)
```

```python
sns.lmplot(x='total_bill', y='tip', data =tips_data, hue='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data =tips_data, col='time')
```
