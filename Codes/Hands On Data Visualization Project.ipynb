```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_churn = pd.read_csv(r'E:\Data Visualization with Python\Datasets\customer_churn.csv')
```

```python
customer_churn.head()
```

```python
data_columns = customer_churn.columns.values.tolist()
print(data_columns)
```

```python
sns.pairplot(data=customer_churn)
```

```python
plt.rcParams["figure.figsize"] = [10,8]
```

```python
sns.barplot(x='Gender', y='Exited', data=customer_churn)
```

```python
plt.title('Age Histogram')
plt.hist(customer_churn["Age"])
```

```python


plt.scatter(customer_churn["Age"], customer_churn["EstimatedSalary"], c = 'g')
```

```python
countries = customer_churn["Geography"].value_counts()

labels = countries.index.values.tolist()
values = countries.values.tolist()
explode = (0.05, 0.05, 0.05) 

plt.pie(values, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.show()
```

```python
sns.boxplot(x='Exited', y='Age', hue = 'Gender',data=customer_churn)
```

```python

sns.violinplot(x='Exited', y='Age', hue = 'Gender',data=customer_churn)
```

```python
sns.set_style('darkgrid')
sns.violinplot(x='Exited', y='Age', hue = 'Gender',data=customer_churn)
```

```python

corr_values = customer_churn.corr()
sns.heatmap(corr_values, annot= True)
```
